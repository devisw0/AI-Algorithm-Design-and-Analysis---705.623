```python
import networkx as nx

my_day = nx.DiGraph()

locations = ["Home", "Office", "Restaurant", "Gym", "Grocery Store"]
my_day.add_nodes_from(locations)

#weight is how long it takes me to get somewhere throughout my day
my_day.add_edge("Home", "Office", weight=60)

# Lunch: Leaving office for food
my_day.add_edge("Office", "Restaurant", weight=15)

#Going home -> wfh rest of the day
my_day.add_edge("Restaurant", "Home", weight=70)

#Workout: Home -> Gym -> Home
my_day.add_edge("Home", "Gym", weight=15)
my_day.add_edge("Gym", "Home", weight=15)

#Groceriy run: Home -> Grocery -> Home
my_day.add_edge("Home", "Grocery Store", weight=10)
my_day.add_edge("Grocery Store", "Home", weight=10)

#Looking at number of nodes and edges
print(f"Graph created with {my_day.number_of_nodes()} locations and {my_day.number_of_edges()} trips.")

#Where I leave from home and go to 
print("\nTrips starting from Home:")
for neighbor in my_day.neighbors("Home"):
    weight = my_day.edges["Home", neighbor]["weight"]
    print(f" -> To {neighbor} ({weight} mins)")

#Where I come form and go hom
print("\nTrips returning to Home:")
for origin in my_day.predecessors("Home"):
    weight = my_day.edges[origin, "Home"]["weight"]
    print(f" <- From {origin} ({weight} mins)")
```
