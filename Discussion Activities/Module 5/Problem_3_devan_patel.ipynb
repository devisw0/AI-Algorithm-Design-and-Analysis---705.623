```python
import random  

def run_simulation(num_samples):

    kept_samples = 0       
    # counts how many samples passed the evidence filter (T=pos AND S=true)
    
    disease_and_evidence = 0  
    # counts how many of those also had D = true (many can pass test and have symptoms but no disease)

    for i in range(num_samples):
        # repeat the sampling process num_samples times

        # number between 0 and 1
        # if that number is less than 0.05, we say D = true (5% chance)
        # gives us a bool based on condition
        D = random.random() < 0.05

        # if D is true then T = positive with 90% chance
        # if D is false, T = positive with only 10% chance
        if D == True:
            T = random.random() < 0.9
        else:
            T = random.random() < 0.1

        # now for Disease and if patient has symptoms
        if D == True:
            S = random.random() < 0.8
        else:
            S = random.random() < 0.2

        # we only care about samples where T = positive and S = true
        if T == True and S == True:
            kept_samples += 1  
            # this sample matched our evidence, so we keep it

            if D == True:
                disease_and_evidence += 1  
                # this kept sample also had disease

    # if no samples were kept, we can't estimate (avoid dividing by zero)
    if kept_samples == 0:
        return None

    posterior = disease_and_evidence / kept_samples
    # out of all kept samples, what fraction had D = true?

    print(f"Samples: {num_samples} | Kept: {kept_samples} | Estimate: {posterior:.4f}")
    # comparing disease true given pops where S is true and T is positive


run_simulation(100)
run_simulation(1000)
run_simulation(10000)
```
