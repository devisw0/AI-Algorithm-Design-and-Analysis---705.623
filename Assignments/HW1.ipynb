## Background on Algorithm Analysis (50 points total)

- Constant Time (O(1)): Regardless of the input size, the algorithm takes a constant amount of time to complete.
    - Accessing an element in an array by index.
    - Inserting or deleting an element at the beginning of a linked list.


- Linear Time (O(n)): The runtime grows linearly with the size of the input.
    - Iterating through a list or array once.
    - Searching for an element in an unsorted list by traversing it.


- Logarithmic Time (O(log n)): As the input size grows, the runtime increases logarithmically.
    - Binary search in a sorted array.
    - Operations in a balanced binary search tree (BST).
    

- Quadratic Time (O(n^2)): The runtime grows quadratically with the input size.
    - Nested loops where each loop iterates through the input.
    - Some sorting algorithms like Bubble Sort or Selection Sort.


- Exponential Time (O(2^n)): The runtime doubles with each addition to the input size.
    - Algorithms involving recursive solutions that make repeated calls.
    - Solving the Traveling Salesman Problem using brute force.
    

- Factorial Time (O(n!)): The runtime grows factorially with the input size.
    - Permutation problems that explore all possible combinations, like the brute force solution for the Traveling Salesman Problem with no optimizations.

## Bubble Sort (Example)

Use the below analysis and explanation as a template for answering the following questions on algorithm analysis. You are to analyze each algorithm line by line, calculate the runtime complexity, and provide an explanation in markdown.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # O(n)
        for j in range(0, n-i-1):  # O(n)
            if arr[j] > arr[j+1]:  # O(1)
                arr[j], arr[j+1] = arr[j+1], arr[j]  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n)
# Big Theta notation: Θ(n^2)
```

Explanation:

- The outer loop iterates 'n' times, where 'n' is the length of the array.
- The inner loop also iterates 'n' times in the worst case.
- The comparisons and swapping operations inside the inner loop are constant time ('O(1)').

This results in a total runtime complexity of O(n^2) in the worst case, Ω(n) in the best case (already sorted), and Θ(n^2) in the average and worst cases, as both loops iterate over the input array.

## 1A Merge Sort (10 points)

```python
def merge_sort(arr):
    if len(arr) > 1:  
        mid = len(arr) // 2  
        left_half = arr[:mid] 
        right_half = arr[mid:]  

        merge_sort(left_half)  
        merge_sort(right_half)  

        i = j = k = 0  

        while i < len(left_half) and j < len(right_half):  
            if left_half[i] < right_half[j]:  
                arr[k] = left_half[i]  
                i += 1
            else:
                arr[k] = right_half[j]  
                j += 1
            k += 1

        while i < len(left_half):  
            arr[k] = left_half[i]  
            i += 1
            k += 1

        while j < len(right_half):  
            arr[k] = right_half[j]  
            j += 1
            k += 1

# Total Runtime Complexity: 
# Big O notation: O(n(log(n)))
# Big Omega notation: Ω(nlog(n))
# Big Theta notation: Θ(nlog(n))
```

Explanation:

- in the merge sort algorithm we are splitting the array into the left half and right half and then using recursion within for each of the halves, continuing until the end point making it O(log(n))
- Then we go through the while loops which is used for itterating through the elements of the arrays we have made each O(n) time
- So in the end we are left with O(nlog(n)) time


## 1B Binary Search (10 points)

```python
def binary_search(arr, target):
#given sorted arrya we want to find a target in the array
#we set a middle value and compare the middle value to the target and adjust low or high depending on it


    low = 0  
    high = len(arr) - 1  

    while low <= high:  
        mid = (low + high) // 2  
        mid_val = arr[mid]  

        if mid_val == target:  
            return mid  
        elif mid_val < target:  
            low = mid + 1 
        else:
            high = mid - 1  

    return -1  

# Total Runtime Complexity:
# Big O notation: O(log n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(log n)
```

Explanation:

- For each iteration we perform a split in the array leading to O(log(n)) time
- In binary search we do this split and then compare the middle value to the target and then shift what the high or low value is depending on comparison with the middle value
- And we then go to another iteration performing another split
- Best case scenario is Ω(1) because there is a chance we our middle is the target value

## 1C Linear Search (10 points)

```python
def linear_search(arr, target):
    for i in range(len(arr)): 
        if arr[i] == target:  
            return i  
    return -1  

# Total Runtime Complexity:
# Big O notation: O(n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(n)
```

Explanation: 

- We have a for loop that checks every value in the array making it O(n)
- from there we only have a conditional and return statement
- Best case scenario or Omega notation would be Ω(1) ib the chance the first array value we check is the target. But for big O and it is O(n) as in the worst case or in between you would have to search through the entire array of n values. 


## 1D Selection Sort (10 points)

```python
def selection_sort(arr):
    #input unsorted array and split into two sections, a sorted and unsorted array
    
    n = len(arr)
    for i in range(n - 1): 
        min_index = i  
        for j in range(i + 1, n):  
            if arr[j] < arr[min_index]:  
                min_index = j  
        arr[i], arr[min_index] = arr[min_index], arr[i]  

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n^2)
# Big Theta notation: Θ(n^2)
```

Explanation:

- We have our outer for loop we use to itterate through n-1 values which would be O(n) 
- Then within that loop we have another for loop going from i+1 to n which is also O(n) making the total run time complexity ->  O(n^2) since thats how many comparisons will always be made

## 1E Heap Sort (10 points)

```python
def heapify(arr, n, i):
    largest = i                      
    left = 2 * i + 1                 
    right = 2 * i + 2             

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  
        heapify(arr, i, 0)               

# Total Runtime Complexity:
# Big O notation: O(n log(n))
# Big Omega notation: Ω(n log(n))
# Big Theta notation: Θ(n log(n))
```

Explanation:

- In the first for loop we are calling heapify within that loop. So the loop worst case itterates through n values making it -> O(n) time.
- In the heapify function that we call in the loop we are sorting a max heap which is O(log n)

- Then once again in the second loop we also have a loop, this one is also O(n) since we are looping through the entries
- And within it we call heapify resulting in O(log(n))

- So this makes the runtime for this search as O(nlog(n))

However, for a dense graph, an adjacency list becomes less space efficient and becomes O(V^2) space. For sparse graphs specifically, an adjacency matrix would be wasting a lot of memory since many of the entries would be 0, and the list would definitely be a better choice. However, in the case the graph is dense, then the matrix would be space efficient because most of the slots are being used to store an edge. For edge look up in a matrix time complexity is O(1) because we just simply go to the row of one of the nodes and to the column of the other node we want to look for the edge for. On the other hand for an adjacency list it would be O(degree of V) (sparse graph) since you would have to go to the node you want to find the edge of and then iterate through the list until you see the other node (if it even is there/exists). Regarding neighbor ittoration the Adjacency List is optimal for sparse graphs as it allows direct access to connected nodes in O(degree(V)) time. The Matrix, however, is not efficient here, as it forces a scan of the entire row, taking O(V) time to identify non-zero entries, resulting in wasted computation for nodes that were never connected. However, in a dense graph O(V) time would be considered efficient because most of the checks would result in finding a neighbor so time is not wasted much.

In an adjacency matrix the runtime is always O(V^2) because we would be visiting every node and would still check the empty spots in both bfs and dfs.
