**Approximate Inference**
- How would you generate samples

Unlike rejection sampling where non-matching samples are thrown away, likelihood weighting guarantees every sample is useful since we always force the evidence variables to their observed values. We make fake patients in topological order, where F and C are randomly sampled from their prior probabilities. When F yes is 0.05 and no is 0.95, C yes is 0.1 and no is 0.9. Fe and Co are forced to yes as our evidence. So the samples we generate would be F: random, C: random, Fe: yes, Co: yes.

- How weights are assigned

From here we would assign weights. Since we forced Fe and Co to yes, we calculate the weight for each sample using the CPT probabilities of the forced evidence variables given whatever values of F and C were randomly sampled for that patient:

w = P(Fe=Yes | F, C) x P(Co=Yes | C)

A higher weight means the sampled F and C values naturally explain the evidence, so that sample counts more. A lower weight means the opposite.


- How the posterior is approximated from weighted samples

After making N samples we compute the posterior:

P(F=Yes | Fe=Yes, Co=Yes) = (sum of weights for samples where F=Yes) / (sum of weights for ALL samples)
So we sum all the weights of all the samples where F is yes and divide by the total weights of all samples. This is an approximation that gets more closer as N increases, converging at 0.1113.
